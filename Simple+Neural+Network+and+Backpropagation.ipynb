```python
#simple neural network 
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#develop array
input = np.arange(-2,2,0.1)
print(input)
```

```python
#calculate n=wp+b for first neuron
n11=-1*input+0.5
print(n11)
n12=1*input+1
print(n12)
```

```python
#create poslin function and apply
n11[np.where(n11<0)]=0
a11=n11

n12[np.where(n12<0)]=0
a12=n12
```

```python
#calculate n=wp+b for second neuron
n21=a11*1+a12*1-1
n21
```

```python
#show plot
plt.plot(input, n21)
plt.show()
```

```python
#calculate n=wp+b
n=1*input+1
print(n)
#apply hardlims function
n[np.where(n<0)]=-1
n[np.where(n>=0)]=1
a=n
print(a)
#plot
plt.plot(input, n)
plt.show()
```
